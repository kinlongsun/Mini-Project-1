```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

from sklearn.linear_model import LinearRegression
```

```python
csv = pd.read_csv('AB_NYC_2019.csv')
csv.head()
```

```python
csv.describe(include='all')
```

```python
# Dropping features that I think are irrelevant

raw_data = csv.drop(['id', 'name', 'host_id', 'host_name', 'last_review', 'calculated_host_listings_count'], axis=1)
raw_data.head()
```

# Missing Variables

```python
# Checking how many NaN we have in my dataset

raw_data.isnull().sum()
```

```python
raw_data.describe(include='all')
```

```python
# Before we fill the Null with the mean of that feature, it came to my mind to check on the nature of the Nulls
# Trying to see what kind of number_of_reviews are corresponding to the reviews_per_month Nulls

raw_data[['number_of_reviews', 'reviews_per_month']][raw_data['reviews_per_month'].isnull()]
```

```python
# Although there are only roughly 10,000 Nulls to check, I realized it isn't always realistic with even larger dataset
# So I thought we can pull the summary statistics that meets the criteria

raw_data[['number_of_reviews']][raw_data['reviews_per_month'].isnull()].describe()
```

```python
# Just as I suspected, the reviews_per_month NaN is due to having no reviews.
# One option is to fill the Nulls with the mean of the feature by using the following code
# raw_data.reviews_per_month = raw_data.reviews_per_month.fillna(raw_data.reviews_per_month.mean())
```

```python
# The option that makes more sense is simply fill them with 0's

data = raw_data
data.reviews_per_month = data.reviews_per_month.fillna(0)
```

```python
# Check again to see if there are anymore Nulls

data.isnull().sum()
```

# Outliers

```python
# We will put the features on the plot and remove some of the problematic samples(may ~.5%)
# We will first investigate the boxplot, or distribution plot, to figure out which side(s) outliers lay on
```

```python
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(18,10))
sns.boxplot(data['price'], ax = ax1)
sns.distplot(data['price'], ax = ax2)
plt.show()
```

```python
# As we can see, it looks like most of the per night cost is under $1,000, and outliers is present.
# The outliers seems to lay on the right side.
# We are going to remove the top 0.5% of the samples. Using logic, an ordinary property shouldn't be charge $10,000 a night

# We want to get the 99.5th percentitle of 'price'
q_price = data['price'].quantile(0.995)

data_1 = data[data['price']<q_price]
```

```python
# Looking at the plots, there are much fewer outliers

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(18,10))
sns.boxplot(data_1['price'], ax = ax1)
sns.distplot(data_1['price'], ax = ax2)
plt.show()
```

```python
# I will do the same for 'minimum_nights', 'number_of_reviews', 'reviews_per_month', and 'availability_365'
```

```python
# Next-'minimum_nights'

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(18,10))
sns.boxplot(data_1['minimum_nights'], ax = ax1)
sns.distplot(data_1['minimum_nights'], ax = ax2)
plt.show()

q_minimum_nights = data_1['minimum_nights'].quantile(0.995)
data_2 = data_1[data_1['minimum_nights']<q_minimum_nights]
```

```python
# After removing some of the more extreme values, it still look like there is a large concentration of minimum night requirement at 30 days
# We can leave this for now 
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(18,10))
sns.boxplot(data_2['minimum_nights'], ax = ax1)
sns.distplot(data_2['minimum_nights'], ax = ax2)
plt.show()
```

```python
# Next-'number_of_reviews'

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(18,10))
sns.boxplot(data_2['number_of_reviews'], ax = ax1)
sns.distplot(data_2['number_of_reviews'], ax = ax2)
plt.show()

q_number_of_reviews = data_2['number_of_reviews'].quantile(0.995)
data_3 = data_2[data_2['number_of_reviews']<q_number_of_reviews]
```

```python
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(18,10))
sns.boxplot(data_3['number_of_reviews'], ax = ax1)
sns.distplot(data_3['number_of_reviews'], ax = ax2)
plt.show()
```

```python
# Next-'reviews_per_month'

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(18,10))
sns.boxplot(data_3['reviews_per_month'], ax = ax1)
sns.distplot(data_3['reviews_per_month'], ax = ax2)
plt.show()

q_reviews_per_month = data_3['reviews_per_month'].quantile(0.995)
data_cleaned = data_3[data_3['reviews_per_month']<q_reviews_per_month]
```

```python
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(18,10))
sns.boxplot(data_cleaned['reviews_per_month'], ax = ax1)
sns.distplot(data_cleaned['reviews_per_month'], ax = ax2)
plt.show()
```
